## Load and join data by county

```python
import pandas as pd
county = pd.read_csv('data/county.csv').drop(['date'], axis=1)
county['death_rate_confirmed'] = county['confirmed_deaths']/county['confirmed_cases']
county['death_rate'] = county['deaths']/county['cases']
county.head()
```

```python
county_mask = pd.read_csv('data/county_mask.csv').drop(['State fips'], axis=1)
county_mask.rename(columns={'COUNTYFP': 'fips'}, inplace=True)
county_mask.head()
```

```python
county_pop = pd.read_csv('data/county_pop.csv').drop(['County', 'STATE'], axis=1)
county_pop.head()
```

```python
df_county = pd.merge(county, county_mask)
df_county = pd.merge(df_county, county_pop)
df_county['comfirmed_cases/population'] = df_county['confirmed_cases']/df_county['Census']
df_county.to_csv('data/aggregated by county.csv', index=False)  
df_county.head()
```

## Load and join data by state

```python
state = pd.read_csv('data/state.csv').drop(['date', 'probable_cases', 'probable_deaths'], axis=1)
state['death_rate_confirmed'] = state['confirmed_deaths']/state['confirmed_cases']
state['death_rate'] = state['deaths']/state['cases']
state.rename(columns={'fips':'State_fips'}, inplace=True)
state.head()
```

```python
state_mask = pd.read_csv('data/state_mask.csv')
state_mask.rename(columns={'State fips': 'State_fips'}, inplace=True)
state_mask.head()
```

```python
state_pop = pd.read_csv('data/state_pop.csv')
state_pop.rename(columns={'State fips': 'State_fips'}, inplace=True)
state_pop.head()
```

```python
df_state = pd.merge(state, state_mask)
df_state = pd.merge(df_state, state_pop)
df_state['cases/population'] = df_state['cases']/df_state['Census']
```

```python
hosp_util = pd.read_csv('data/hosp_util.csv').drop(['OBJECTID'], axis=1)
hosp_util.head()
```

```python
state_insurance = pd.read_csv('data/state_insurance.csv').drop(['State'], axis=1)
state_insurance.head()
```

```python
df_state = pd.merge(df_state, hosp_util)
df_state = pd.merge(df_state, state_insurance)
df_state['Any_coverage/population'] = df_state['Any coverage']/df_state['Census']
df_state.to_csv('data/aggregated by state.csv', index=False)  
df_state.head()
```
